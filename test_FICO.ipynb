```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from igann import IGANN, IGANNClassifier


np.random.seed(42)

from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
data_path = "/Users/JudyWork/Desktop/Research/474-GAMS/data/FICO/heloc_dataset_v1.csv"
df = pd.read_csv(data_path)

X = df.drop('RiskPerformance', axis=1).astype(np.float64)
y = (df['RiskPerformance'] == 'Good').astype(int) # Convert to binary: 1 for 'Good', 0 for 'Bad'

# split into 15% test, 10% validation, 75% train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1/(1-.15), random_state=43)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [10, 30, 50, 100, 200, 500, 1000],
    'boost_rate': [0.01, 0.03, 0.05, 0.1, 0.2, 0.5],
    'n_hid': [5, 10, 20, 30, 50, 70]
}

# scoring = ['accuracy','roc_auc']

igann_classifier = IGANNClassifier(task='classification')

grid_search = GridSearchCV(igann_classifier, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)
print(grid_search.cv_results_)
```

```python
# Fit model

igann = IGANN(task='classification', n_estimators=30, boost_rate=0.5, n_hid=20)
igann.fit(X_train, y_train)

# Display results

y_pred = igann.predict(X_test)

test_accuracy = accuracy_score(y_test, y_pred)
test_auc = roc_auc_score(y_test, y_pred)


print(f"Accuracy: {test_accuracy}, AUC: {test_auc}")

igann.plot_single(show_n=23)
```

```python
n_estimators_list = [10, 20, 30, 40, 50, 100, 200, 500, 1000]

fit_times = {}
models = {}

for n_estimators in n_estimators_list:
    print(f"\nFitting GAM with {n_estimators} estimators...")
    start_time = time.time()
    
    # Fit model
    igann = IGANN(task='classification', n_estimators=n_estimators, boost_rate=0.01)
    igann.fit(X_train, y_train)
    
    # Measure fitting time
    end_time = time.time()
    fit_times[n_estimators] = end_time - start_time
    models[n_estimators] = igann 
    
    # Display results
   
    y_pred = igann.predict(X_test)
    
    test_accuracy = accuracy_score(y_test, y_pred)
    test_auc = roc_auc_score(y_test, y_pred)
    

    print(f"Accuracy: {test_accuracy}, AUC: {test_auc}")
    
    igann.plot_single(show_n=23)
       

# Print fitting times
print("\nFitting Times (seconds):")
for n, t in fit_times.items():
    print(f"{n} smoothing parameters: {fit_times[n]:.4f} sec")

# Plot fitting times separately
plt.figure(figsize=(8, 5))
plt.plot(n_estimators_list, [fit_times[n] for n in n_estimators_list], marker='o')
plt.xlabel('Number of Estimators')
plt.ylabel('Fitting Time (seconds)')
plt.title('Fitting Time vs. Number of Estimators')
plt.grid(True)
plt.savefig('IGANN_gam_timing_comparison.png', dpi=300)
plt.show()
```
